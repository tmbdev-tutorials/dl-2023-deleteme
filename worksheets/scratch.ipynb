```python
import numpy as np
```

```python
v1 = np.array([1.0, 2.0])
v1 /= np.linalg.norm(v1)
v2 = np.array([3.0, 1.0])
v2 /= np.linalg.norm(v2)
```

```python
n = 10000
v1s = np.random.uniform(-1, 1, (n,1)) @ v1.reshape(1,2)
v2s = np.random.uniform(-1, 1, (n,1)) @ v2.reshape(1,2)
figsize(12, 5)
subplot(121)
scatter(v1s[:,0], v1s[:,1], marker=".", alpha=0.3)
scatter(v2s[:,0], v2s[:,1], marker=".", alpha=0.3)
subplot(122)
s = v1s + v2s
scatter(s[:,0], s[:,1], marker=".", alpha=0.3)
```

```python
n = 10000
v1s = np.random.normal(size=(n,1)) @ v1.reshape(1,2)
v2s = np.random.normal(size=(n,1)) @ v2.reshape(1,2)
figsize(12, 5)
subplot(121)
scatter(v1s[:,0], v1s[:,1], marker=".", alpha=0.3)
scatter(v2s[:,0], v2s[:,1], marker=".", alpha=0.3)
subplot(122)
s = v1s + v2s
scatter(s[:,0], s[:,1], marker=".", alpha=0.3)
```

```python
import re
shakespeare = open("t8.shakespeare.txt").read()
shakespeare = re.sub("'s( |$)", " ", shakespeare)
shakespeare = re.sub(r"[^a-zA-Z]+", " ", shakespeare)
shakespeare = shakespeare.lower().split()
len(shakespeare)

from collections import Counter
bigram = {}
for i in range(len(shakespeare)-1):
    word = shakespeare[i]
    bigram.setdefault(shakespeare[i], Counter())[shakespeare[i+1]] += 1
bigram["the"].most_common(10)
```

```python
bigram["king"].most_common(10)
```

```python
n = sum(bigram["king"].values())
```

```python
for k, v in bigram["king"].most_common(10):
    print("king %10s %.4f" % (k[:10], v/n))
```

```python
n = 3
ngrams = {}
for i in range(len(shakespeare)-n):
    context = tuple(shakespeare[i:i+n-1])
    ngrams.setdefault(context, Counter())[shakespeare[i+n-1]] += 1
```
